```python
import pandas as pd
import statistics
import csv
df=pd.read_csv("StudentsPerformance.csv")
Score=df["reading score"].to_list()
meanSc=statistics.mean(Score)
medianSc=statistics.median(Score)
modeSc=statistics.mode(Score)
stdSc=statistics.stdev(Score)

std1ScStart,std1ScEnd=meanSc-stdSc,meanSc+stdSc
std2ScStart,std2ScEnd=meanSc-(2*stdSc),meanSc+(2*stdSc)
std3ScStart,std3ScEnd=meanSc-(3*stdSc),meanSc+(3*stdSc)


ScoreWithinStd1=[result for result in Score
                 if result>std1ScStart and result<std1ScEnd

                 ]

ScoreWithinStd2=[result for result in Score
                 if result>std2ScStart and result<std2ScEnd
                 
                 ]


ScoreWithinStd3=[result for result in Score
                 if result>std3ScStart and result<std3ScEnd
                 
                 ]


print("mean,mode,median of Score is{},{},{}".format(meanSc,modeSc,medianSc))



print("{}% of  Data for Score within 1sd".format(len(ScoreWithinStd1)*100.0/len(Score)))
print("{}% of  Data for Score within 2sd".format(len(ScoreWithinStd2)*100.0/len(Score)))
print("{}% of  Data for Score within 3sd".format(len(ScoreWithinStd3)*100.0/len(Score)))




```
